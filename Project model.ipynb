```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("train_co.csv",index_col=False)
df_train.head()
```

```python
labels=df_train["784"]
```

```python
df_train.drop(df_train.columns[[784]],axis=1,inplace=True)
df_train.head()
```

```python
np.random.seed(1212)
import keras
from keras.models import Model
from keras.layers import *
from keras import optimizers
from keras.layers.convolutional import Conv2D, MaxPooling2D 
from keras.models import Sequential
from keras.utils import np_utils
```

```python
labels=np.array(labels)
```

```python
from keras.utils.np_utils import to_categorical
cat = to_categorical(labels,num_classes=13)
```

```python
l=[]
for i in range(7567):
    l.append(np.array(df_train[i:i+1]).reshape(28,28,1))
```

```python
np.random.seed(7)
```

```python
model=Sequential()
model.add(Conv2D(512,(5,5),input_shape=(28,28,1),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(256,(3,3),input_shape=(28,28,1),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(128,(3,3),input_shape=(28,28,1),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128,activation="relu"))
model.add(Dense(64,activation="relu"))
model.add(Dense(32,activation="relu"))
model.add(Dense(13,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(np.array(l),cat,shuffle=True,epochs=10)
```
